```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate sample data or load your dataset
# For illustration purposes, let's create random data
np.random.seed(42)
mortgage_rates = np.random.uniform(low=3.0, high=5.0, size=100)
home_values = 300000 + 50000 * mortgage_rates + np.random.normal(scale=10000, size=100)

# Create a DataFrame with the data
data = pd.DataFrame({'Mortgage_Rates': mortgage_rates, 'Home_Values': home_values})

# Visualize the data
plt.scatter(data['Mortgage_Rates'], data['Home_Values'])
plt.title('Mortgage Rates vs. Home Values')
plt.xlabel('Mortgage Rates (%)')
plt.ylabel('Home Values ($)')
plt.show()

# Prepare the data for linear regression
X = data['Mortgage_Rates'].values.reshape(-1, 1)
y = data['Home_Values'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print the regression equation
print(f'Regression Equation: Home Values = {intercept:.2f} + {slope:.2f} * Mortgage Rates')

# Visualize the linear regression line
plt.scatter(data['Mortgage_Rates'], data['Home_Values'])
plt.plot(data['Mortgage_Rates'], model.predict(X), color='red', linewidth=2)
plt.title('Mortgage Rates vs. Home Values with Linear Regression')
plt.xlabel('Mortgage Rates (%)')
plt.ylabel('Home Values ($)')
plt.show()
```
